# Bepaling van soortelijke warmte van water


# Introductie

Water heeft een enorme buffercapaciteit: je moet veel energie toevoegen om de temperatuur van water een graad te verwarmen. In dit practicum gaan we de soortelijke warmte van water bepalen door een bekende hoeveelheid water te verwarmen met een bekende hoeveelheid energie, en de temperatuurstijging te meten.

# Theorie

Geef hier de natuurkundige achtergrond van het onderwerp. 

````python
Soortelijke warmte is een grootheid uitgedrukt in de eenheiden: J/(kg*K), dit laat zien hoeveel energie er nodig is om 1 kilogram van een bepaalde stof 1 kelvin te verwarmen.
Om de soortelijke warmte van water te bepalen moet er dus gemeten worden hoeveel graden kelvin een bepaalde hoeveelheid water stijgt na toevoer van een bekende hoeveelheid energie (in de vorm van warmte).

De massa/hoeveelheid water kan bepaalt worden door water eerst in een maatbeker te gieten.
Doormiddel van een weerstand onder een bepaalt vermogen met een gemeten tijd, kan precies worden bepaalt hoeveel energie er in het systeem wordt gebracht.
Het enige wat nu nog gemeten moet worden is de temperatuur verschil voor en na het inbrengen van de weerstand.

door deze forumle in te vullen kan de soortelijkewarmte van water berekend worden
c = Q / (m * dT)
````
